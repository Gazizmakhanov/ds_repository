```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express
from sklearn.experimental import enable_iterative_imputer

from sklearn.impute import IterativeImputer
```

```python
df = pd.read_csv('vehicles_us.csv')
df.head()
```

```python
plt.scatter(df['cylinders'], df['price'])

plt.xlabel('cylinders')
plt.ylabel('price')
plt.title('cylinder to price')
plt.show()
```

Price for six and eight cylinder cars are the highest and for 12 cylinders being the lowest

```python
plt.hist(df['condition'], bins=10, edgecolor='black')
plt.xticks(rotation=90, fontsize=10)

plt.xlabel('Condition')
plt.ylabel('Frequency')
plt.title('Histogram of the car conditions ')

# Show plot
plt.show()
```

Majority of the cars are in good and excellent conditions. The least amount of cars being salvage and new.

```python
counts, bins, patches = plt.hist(df['type'], bins=10, edgecolor='black')

# Assign different colors to each bin
colors = ['red', 'green', 'blue', 'orange', 'purple']
for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

# Labels and formatting
plt.xlabel('Types')
plt.ylabel('Frequency')
plt.title('Histogram of car types')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()

# Show plot
plt.show()
```

From the chart above we can conclude that SUVs and pickup are the most popular cars types. Buses, offroads and mini-vans are among the least popular ones.
